## Import dados usando a biblioteca pandas

```python
import pandas as pd
```

```python
data = pd.read_excel("plinc.xlsx", skiprows=1)
```

```python
data.columns # acessar a coluna 
```

```python
data["Tempo"].values
```

```python
data
```

```python
data2 = data[["Tempo", "Distance [mm]"]] # obter uma nova dataframe
```

```python
data2["Tempo_s"] = data2["Tempo"]/1000
data2['Distance[m]'] = data2["Distance [mm]"]/1000
```

```python
data2
```

```python
import matplotlib.pyplot as plt
```

```python
x = data2["Tempo_s"]
y = data2["Distance[m]"]

fig, ax = plt.subplots(figsize=(8,5))
plt.plot(x,y)
plt.xlabel("Tempo(s)", fontsize=15)
plt.ylabel("Distance(m)", fontsize=15)
plt.grid()

ax.tick_params(axis='both', which='major', labelsize=12)
plt.savefig('plano_inclinado_resultado' + '.pdf', format='pdf', dpi=300, bbox_inches='tight')
```
